## Exploratory data analysis

boxplots

```r
options(repr.plot.width = 8, repr.plot.height = 5)
#tiff('Boxplot.tiff',width=8,height=5,units='in',res=300)
p <- ggplot(df) +
geom_boxplot(aes(x=Treatment, y=AUDPC, color=Treatment)) +
geom_jitter(aes(x=Treatment, y=AUDPC, color=Treatment),
            alpha=0.8,
            position=position_jitter(width=0.1)) + 
labs(y="Early blight disease intensity (AUSPC)") +
theme_classic() + theme(legend.position = "none")

# Marginal histogram
mh <- ggplot(df, aes(x=AUDPC, fill=Treatment)) + 
  geom_density(alpha = 0.4) + 
  theme_void() + 
  theme(legend.position = "none") + 
  coord_flip()

# Plot all
p + mh + plot_layout(ncol = 2, nrow = 1,
                    widths = c(4, 1), heights = c(1,1))

#dev.off()
```

## Inference

1. Fit ANOVA model

```r
m <- lm(AUDPC ~ Treatment + Block,
       data=df)
m2 <- aov(AUDPC ~ Treatment + Block,
       data=df)
```

```r
lm <- lm(log(AUDPC) ~ Treatment + Block,
       data=df)
lm2 <- aov(log(AUDPC) ~ Treatment + Block,
       data=df)
```

- summaries

```r
anova(m)
```

```r
summary(m)
```

## Diagnostics

```r
plot(m, col=df$Treatment, pch=20, cex=2)
```

#### Hypothesis test for normality

```r
shapiro.test(residuals(m2))
```

#### Hypothesis test for equal variances

```r
leveneTest(log(AUDPC) ~ Treatment, data = df)
```

```r
leveneTest(AUDPC ~ Treatment, data = df)
```

```r
fligner.test(AUDPC ~ Treatment, data = df)
```

## Randomization Inference

#### Declare the randomization procedure

```r
blocks = df$Block #extract the block column from the original dataset
declaration <- declare_ra(N = 64, num_arms = 16, blocks = blocks)
declaration
```

#### Assign the treatments randomly

```r
Z = conduct_ra(declaration)
length(Z)
unique(Z)
df= cbind(df, Z)# join randomly assigned treatment
```

```r
blocks = df$Block #extract the block column from the original dataset
declaration <- declare_ra(N = 64, num_arms = 16, blocks = blocks)
declaration
Z = conduct_ra(declaration)
length(unique(Z))
df= cbind(df, Z)
ri2_out <-
  conduct_ri(
    model_1 = AUDPC ~ 1, # restricted model
    model_2 = AUDPC ~ Z, # unrestricted model
    declaration = declaration,
    sharp_hypothesis = 0,
    data = df
  )

plot(ri2_out)
```

```r
summary(ri2_out)
```
